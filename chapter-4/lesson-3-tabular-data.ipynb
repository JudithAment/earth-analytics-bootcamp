```python
# Import packages
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Download and open the .csv file using Pandas
avg_monthly_precip = pd.read_csv(
    "https://ndownloader.figshare.com/files/12710618"
)
```

```python
# View the data that you just downloaded and opened
avg_monthly_precip
avg_monthly_precip.info()
```

```python
# Plot the data
fig, ax = plt.subplots()
avg_monthly_precip.plot(
    x="months",
    y="precip",
    title="Precipitation (mm) for one year",
    ax=ax
)
ax.set(
    xlabel='Month',
    ylabel='Precipitation (mm)'
)
```

```python
# Challenge 2
# Make it a bar plot
fig, ax = plt.subplots()
avg_monthly_precip.plot(
    x="months",
    y="precip",
    title="Precipitation (mm) for one year",
    color="green",
    kind="bar",
    ax=ax
)

ax.set(
    xlabel='Month',
    ylabel='Precipitation (mm)'
)
```

```python
# Challenge 3
# Open temperature data for Miami, Florida
miami_temp_url = "https://www.ncdc.noaa.gov/cag/city/time-series/USW00012839-tmax-12-12-1895-2020.csv"

# Open the Miami data skipping the first 3 rows and setting no data values
miami_temp = pd.read_csv(
    miami_temp_url,
    skiprows=3,
    na_values=-99
)

# View the first 5 rows of the data
miami_temp.head()
```

```python
# Plot miami data
fig, ax = plt.subplots()
miami_temp.plot(
    x="Date",
    y="Value",
    title="Temperature (F) for Miami, FL",
    ax=ax
)
ax.set(
    xlabel='Date (yyyymm)',
    ylabel='Temperature (F)'
)
```

```python
# Challenge 4
# Get Seattle data
seattle_url = "https://www.ncdc.noaa.gov/cag/city/time-series/USW00013895-tmax-1-5-1895-2020.csv"
seattle_temp = pd.read_csv(
    seattle_url,
    skiprows=3,
    na_values=-99
)

seattle_temp.info()
seattle_temp.head()

# Plot seattle data
fig, ax = plt.subplots()
seattle_temp.plot(
    x="Date",
    y="Value",
    title="Challenge 4: Temperature (F) for Montgomery, AL",
    ax=ax
)
ax.set(
    xlabel='Date (yyyymm)',
    ylabel='Temperature (F)'
)
plt.show()
```

```python
# Challenge 5
# Download greenhouse gas CO2 data
import urllib.request

greenhouse_gas_url = "ftp://aftp.cmdl.noaa.gov/data/trace_gases/co2/in-situ/surface/brw/co2_brw_surface-insitu_1_ccgg_MonthlyData.txt"

urllib.request.urlretrieve(url=greenhouse_gas_url, 
                           filename="co2-emissions-barrow.csv")

barrow_data = pd.read_csv("co2-emissions-barrow.csv", sep=' ', skiprows=149, na_values=-999.99)
barrow_data.info()
barrow_data.head()

fig2, ax = plt.subplots()
barrow_data.plot(x="time_decimal",
                  y="value",
                  title="Challenge 5: CO2 in Utqiagvik",
                  ax=ax)
ax.set(
    xlabel='Date',
    ylabel='CO2'
)
plt.show()
```
