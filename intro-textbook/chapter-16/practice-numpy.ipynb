## Practice NumPy skills

```python
import numpy as np
import earthpy as et
```

### Challenge 1

```python
months_url = 'https://ndownloader.figshare.com/files/12565619'
files_path = et.data.get_data(url = months_url)
months = np.genfromtxt(files_path, dtype = "str")
```

### Challenge 2

```python
# Download data from URL to .csv of precip data for 2002 and 2013
precip_2002_2013_url = 'https://ndownloader.figshare.com/files/12707792'
files_path = et.data.get_data(url=precip_2002_2013_url)
precip_2002_2013 = np.loadtxt(files_path, delimiter = ",")
```

```python
precip_2013 = precip_2002_2013[1, :]
print(precip_2013)
```

### Challenge 3

```python
precip_2013_mm = precip_2013 * 25.4
```

### Challenge 4

```python
import matplotlib.pyplot as plt

plt.plot(
    months,
    precip_2013_mm
)

plt.title("Average Monthly Precipitation\nBoulder, CO")
plt.ylabel("Precipitation (mm)")
```
